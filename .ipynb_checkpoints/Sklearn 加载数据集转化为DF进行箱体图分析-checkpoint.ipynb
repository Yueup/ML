## 基于Scikit-learn对波士顿数据集进行箱体图，散点图分析

```python
from sklearn.datasets import load_boston
boston=load_boston() #导入数据集
#print(boston.DESCR)

feature=boston.data  #特征
label=boston.target  #标签
fea_nm=boston.feature_names #每个特征的名字

df_fea=pd.DataFrame(feature,columns=(fea_nm))  #创建特征的DF

df_lab=pd.DataFrame(label,columns=['MEDV'])    #创建标签的DF

df=pd.concat([df_fea,df_lab],axis=1,join='inner') #合并，得到整个数据集的DF

df

df.boxplot(column=['RM']) #对RM特征进行箱体图分析


df.plot(kind = "scatter",x = 'RM' ,y = 'DIS' ,s = 30,figsize = (10,6)) #对两个特征之间进行散点图分析
```
