#### 1.两个变量
检验两个变量之间是否独立 

##### 卡方检验  
若P值远小于0.05，则变量不是独立的

```python
# tab2=pd.crosstab(titanic.Embarked,[titanic.Survived,titanic.Sex]) #第一个是行，第二个列 
# tab2
from scipy.stats import chi2_contingency #导入卡方检验方法

tab=pd.crosstab(titanic.Embarked,titanic.Survived)
tab
tab.values

c,p,_,_= chi2_contingency(tab.values)
print("卡方值 : ", c, ";  P value : ",p)
```

##### 马赛克图

```python
# 用于分析两个变量之间的独立性
from statsmodels.graphics.mosaicplot import mosaic

mosaic(titanic,["Embarked","Survived"],gap=0.01, title = "马赛克图")
```
